````python
import pandas as pd

def load_data(file_path):
    return pd.read_csv(file_path)

def clean_data(df):
    df.drop_duplicates(inplace=True)
    df.ffill(inplace=True)
    return df

def normalize_data(df):
    if 'price' in df.columns:
        df['price'] = (df['price'] - df['price'].min()) / (df['price'].max() - df['price'].min())
    return df

def extract_features(df):
    new_columns = pd.DataFrame({
        'quantity': [1] * len(df),
        'total_spent': df['price'] * 1 if 'price' in df.columns else [0] * len(df)
    })
    df = pd.concat([df, new_columns], axis=1)
    return df
````

### Implementa√ß√£o do Algoritmo Gen√©tico

**Genetic Algorithm:**

````python
!pip install deap

import random
from deap import base, creator, tools, algorithms
import numpy as np

def simulated_conversion(product):
    return random.uniform(0, 1)

def simulated_purchase_value(product):
    return random.uniform(10, 100)

def simulated_rejection(product):
    return random.uniform(0, 1)

def evaluate(individual):
    conversion_rate = np.mean([simulated_conversion(product) for product in individual])
    avg_purchase_value = np.mean([simulated_purchase_value(product) for product in individual])
    rejection_rate = np.mean([simulated_rejection(product) for product in individual])
    fitness = conversion_rate * 0.7 + avg_purchase_value * 0.2 - rejection_rate * 0.1
    return (fitness,)

def initialize_population(pop_size, num_products, product_ids):
    population = []
    for _ in range(pop_size):
        individual = random.sample(list(product_ids), num_products)
        population.append(creator.Individual(individual))
    return population

def setup_ga(population, toolbox):
    toolbox.register("mate", tools.cxOnePoint)
    toolbox.register("mutate", tools.mutShuffleIndexes, indpb=0.05)
    toolbox.register("select", tools.selTournament, tournsize=3)
    toolbox.register("evaluate", evaluate)

    population = list(map(toolbox.clone, population))
    algorithms.eaSimple(population, toolbox, cxpb=0.7, mutpb=0.2, ngen=50, stats=None, halloffame=None, verbose=True)
    return population

def main(processed_data):
    pop_size = 100
    num_products = 10
    product_ids = processed_data['product_id'].unique()

    if not hasattr(creator, "FitnessMax"):
        creator.create("FitnessMax", base.Fitness, weights=(1.0,))
    if not hasattr(creator, "Individual"):
        creator.create("Individual", list, fitness=creator.FitnessMax)

    toolbox = base.Toolbox()
    toolbox.register("individual", tools.initIterate, creator.Individual, random.sample)
    toolbox.register("population", tools.initRepeat, list, toolbox.individual)

    population = initialize_population(pop_size, num_products, product_ids)
    final_population = setup_ga(population, toolbox)

    ga_results = []
    metrics = {'conversion_rates': [], 'avg_purchase_values': [], 'rejection_rates': []}

    for ind in final_population:
        fitness = evaluate(ind)[0]
        metrics['conversion_rates'].append(np.mean([simulated_conversion(product) for product in ind]))
        metrics['avg_purchase_values'].append(np.mean([simulated_purchase_value(product) for product in ind]))
        metrics['rejection_rates'].append(np.mean([simulated_rejection(product) for product in ind]))
        ga_results.append(fitness)

    return final_population, ga_results, metrics
````

### Avalia√ß√£o e Compara√ß√£o

**Evaluation:**

````python
def calculate_average_metrics(metrics):
    avg_conversion_rate = np.mean(metrics['conversion_rates'])
    avg_purchase_value = np.mean(metrics['avg_purchase_values'])
    avg_rejection_rate = np.mean(metrics['rejection_rates'])

    return avg_conversion_rate, avg_purchase_value, avg_rejection_rate
````

````python
import pandas as pd
import matplotlib.pyplot as plt
import random
import numpy as np

def baseline_evaluation(processed_data):
    product_ids = processed_data['product_id'].unique()
    baseline_results = []
    for _ in range(50):
        individual = random.sample(list(product_ids), 10)
        fitness = evaluate(individual)[0]
        baseline_results.append(fitness)
    return baseline_results

def compare_with_baseline(ga_results, baseline_results):
    min_length = min(len(ga_results), len(baseline_results))
    ga_results = ga_results[:min_length]
    baseline_results = baseline_results[:min_length]

    data = {
        'GA Results': ga_results,
        'Baseline Results': baseline_results
    }
    df = pd.DataFrame(data)
    return df

def plot_results(data):
    plt.figure(figsize=(10, 5))
    plt.plot(data['GA Results'], label='GA Results', color='b')
    plt.plot(data['Baseline Results'], label='Baseline Results', color='r')
    plt.xlabel('Iteration')
    plt.ylabel('Value')
    plt.title('Comparison of GA and Baseline Results')
    plt.legend()
    plt.show()
````

### Fluxo de Execu√ß√£o

**Main Execution:**

````python
# Carregar e pr√©-processar os dados
processed_data = load_data('/content/drive/MyDrive/Pos tech/fase 2/dataset.csv')
processed_data = clean_data(processed_data)
processed_data = normalize_data(processed_data)
processed_data = extract_features(processed_data)

# Salvar os dados processados
processed_data.to_csv('/content/processed_data.csv', index=False)

# Executar o algoritmo gen√©tico
final_population, ga_results, metrics = main(processed_data)

# Calcular m√©tricas m√©dias do GA
avg_conversion_rate, avg_purchase_value, avg_rejection_rate = calculate_average_metrics(metrics)

# Exibir resultados do algoritmo gen√©tico
print(f"Resultados do Algoritmo Gen√©tico:")
print(f"Taxa de Convers√£o: {avg_conversion_rate:.2f}")
print(f"Valor M√©dio das Compras: ${avg_purchase_value:.2f}")
print(f"Taxa de Rejei√ß√£o: {avg_rejection_rate:.2f}")

# Executar a linha de base
baseline_results = baseline_evaluation(processed_data)

# Comparar e plotar os resultados
data = compare_with_baseline(ga_results, baseline_results)
plot_results(data)
````
