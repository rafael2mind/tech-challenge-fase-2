4. **Divis√£o de Dados**:
   - Separa√ß√£o dos dados em conjuntos de treinamento e teste, garantindo uma avalia√ß√£o justa do modelo.

#### Implementa√ß√£o do Algoritmo Gen√©tico

##### Defini√ß√£o da Popula√ß√£o Inicial

A popula√ß√£o inicial consiste em um conjunto de solu√ß√µes potenciais, cada uma representando uma lista de produtos recomendados para um usu√°rio. As listas podem ser geradas aleatoriamente ou utilizando heur√≠sticas baseadas em dados hist√≥ricos de compras.

1. **Representa√ß√£o dos Indiv√≠duos**:
   - Cada indiv√≠duo √© representado por um vetor de IDs de produtos.
   - Exemplo: [101, 203, 305, 409, 512], onde cada n√∫mero representa um produto espec√≠fico.

2. **Tamanho da Popula√ß√£o**:
   - A popula√ß√£o inicial deve ser suficientemente grande para explorar uma ampla gama de solu√ß√µes, mas pequena o bastante para permitir uma computa√ß√£o eficiente.
   - Exemplo: Uma popula√ß√£o inicial de 100 indiv√≠duos.

##### Fun√ß√£o de Avalia√ß√£o (Fitness Function)

A fun√ß√£o de avalia√ß√£o mede a qualidade de cada lista de recomenda√ß√£o com base em v√°rios crit√©rios de desempenho. Para este projeto, a fun√ß√£o de avalia√ß√£o considerar√°:

1. **Taxa de Convers√£o**:
   - Propor√ß√£o de produtos recomendados que resultam em compras.
2. **Valor M√©dio das Compras**:
   - Valor m√©dio das transa√ß√µes resultantes das recomenda√ß√µes.
3. **Taxa de Rejei√ß√£o**:
   - Propor√ß√£o de produtos recomendados que s√£o ignorados ou rejeitados pelos usu√°rios.

A f√≥rmula da fun√ß√£o de avalia√ß√£o √©:



```
Fitness = w1 * Taxa de Convers√£o + w2 * Valor M√©dio das Compras - w3 * Taxa de Rejei√ß√£o
```
Onde `w1`, `w2` e `w3` s√£o pesos que refletem a import√¢ncia relativa de cada m√©trica.


##### Operadores Gen√©ticos

1. **Sele√ß√£o**:
   - **M√©todo**: Sele√ß√£o por torneio.
   - **Descri√ß√£o**: Um conjunto de indiv√≠duos √© selecionado aleatoriamente, e o melhor indiv√≠duo dentro deste conjunto √© escolhido para reprodu√ß√£o.
   - **Objetivo**: Garantir que os melhores indiv√≠duos tenham maior probabilidade de passar seus genes para a pr√≥xima gera√ß√£o.

2. **Cruzamento (Crossover)**:
   - **M√©todo**: Crossover de um ponto.
   - **Descri√ß√£o**: Dois indiv√≠duos (pais) s√£o combinados, trocando segmentos de seus vetores de recomenda√ß√£o a partir de um ponto aleat√≥rio.
   - **Objetivo**: Introduzir diversidade na popula√ß√£o e explorar novas combina√ß√µes de produtos.
   - **Exemplo**:
     - Pai 1: [101, 203, 305, 409, 512]
     - Pai 2: [102, 204, 306, 408, 511]
     - Filhos: [101, 203, 305, 408, 511] e [102, 204, 306, 409, 512]

3. **Muta√ß√£o**:
   - **M√©todo**: Muta√ß√£o aleat√≥ria.
   - **Descri√ß√£o**: Um ou mais genes de um indiv√≠duo s√£o alterados aleatoriamente.
   - **Objetivo**: Introduzir novas varia√ß√µes e evitar a converg√™ncia prematura.
   - **Exemplo**: [101, 203, 305, 409, 512] -> [101, 207, 305, 409, 512]

##### Crit√©rios de Parada

Os crit√©rios de parada determinam quando o algoritmo deve interromper a evolu√ß√£o. Para este projeto, consideramos os seguintes crit√©rios:

1. **N√∫mero M√°ximo de Gera√ß√µes**:
   - O algoritmo para ap√≥s um n√∫mero pr√©-determinado de gera√ß√µes.

### Carregamento e Pr√©-processamento de Dados

**Data Preprocessing:**

````python
import pandas as pd
import numpy as np

def load_data(file_path):
    return pd.read_csv(file_path)

def clean_data(df):
    df.drop_duplicates(inplace=True)
    df.fillna(method='ffill', inplace=True)
    return df

def normalize_data(df):
    if 'price' in df.columns:
        df['price'] = (df['price'] - df['price'].min()) / (df['price'].max() - df['price'].min())
    return df

def extract_features(df):
    new_columns = pd.DataFrame({
        'quantity': [1] * len(df),
        'total_spent': df['price'] * 1 if 'price' in df.columns else [0] * len(df)
    })
    df = pd.concat([df, new_columns], axis=1)
    return df
````

### Implementa√ß√£o do Algoritmo Gen√©tico

**Genetic Algorithm:**

````python
!pip install deap

import random
from deap import base, creator, tools, algorithms
import numpy as np
import pandas as pd

def simulated_conversion(product):
    return random.uniform(0, 1)

def simulated_purchase_value(product):
    return random.uniform(10, 100)

def simulated_rejection(product):
    return random.uniform(0, 1)

def evaluate(individual):
    conversion_rate = np.mean([simulated_conversion(product) for product in individual])
    avg_purchase_value = np.mean([simulated_purchase_value(product) for product in individual])
    rejection_rate = np.mean([simulated_rejection(product) for product in individual])
    return conversion_rate * 0.7 + avg_purchase_value * 0.2 - rejection_rate * 0.1,

def initialize_population(pop_size, num_products, product_ids):
    population = []
    for _ in range(pop_size):
        individual = random.sample(list(product_ids), num_products)
        population.append(creator.Individual(individual))
    return population

def setup_ga(population, toolbox):
    toolbox.register("mate", tools.cxOnePoint)
    toolbox.register("mutate", tools.mutShuffleIndexes, indpb=0.05)
    toolbox.register("select", tools.selTournament, tournsize=3)
    toolbox.register("evaluate", evaluate)

    population = list(map(toolbox.clone, population))
    algorithms.eaSimple(population, toolbox, cxpb=0.7, mutpb=0.2, ngen=50, stats=None, halloffame=None, verbose=True)
    return population

def main(processed_data):
    pop_size = 100
    num_products = 10
    product_ids = processed_data['product_id'].unique()

    creator.create("FitnessMax", base.Fitness, weights=(1.0,))
    creator.create("Individual", list, fitness=creator.FitnessMax)

    toolbox = base.Toolbox()
    toolbox.register("individual", tools.initIterate, creator.Individual, random.sample)
    toolbox.register("population", tools.initRepeat, list, toolbox.individual)

    population = initialize_population(pop_size, num_products, product_ids)
    final_population = setup_ga(population, toolbox)

    ga_results = [ind.fitness.values[0] for ind in final_population]
    return final_population, ga_results
````

### Avalia√ß√£o e Compara√ß√£o

**Evaluation:**

````python
import pandas as pd
import matplotlib.pyplot as plt
import random
import numpy as np

def baseline_evaluation(processed_data):
    product_ids = processed_data['product_id'].unique()
    baseline_results = []
    for _ in range(50):
        individual = random.sample(list(product_ids), 10)
        fitness = evaluate(individual)[0]
        baseline_results.append(fitness)
    return baseline_results

def compare_with_baseline(ga_results, baseline_results):
    min_length = min(len(ga_results), len(baseline_results))
    ga_results = ga_results[:min_length]
    baseline_results = baseline_results[:min_length]

    data = {
        'GA Results': ga_results,
        'Baseline Results': baseline_results
    }
    df = pd.DataFrame(data)
    return df

def plot_results(data):
    plt.figure(figsize=(10, 5))
    plt.plot(data['GA Results'], label='GA Results', color='b')
    plt.plot(data['Baseline Results'], label='Baseline Results', color='r')
    plt.xlabel('Iteration')
    plt.ylabel('Value')
    plt.title('Comparison of GA and Baseline Results')
    plt.legend()
    plt.show()
````
